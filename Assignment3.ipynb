1. Basic Line Plot                                                               
o Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.        
o Label the x-axis as "X Values" and the y-axis as "Squared Values."             
o Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10, 11, 1)
y = x**2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Graph of y = x^2")

# Show the plot
plt.show()
```

2. Bar Plot                                                                      
o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',         
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].                        
o Add labels to the x-axis and y-axis and give the chart a title.

```python
import matplotlib.pyplot as plt
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

# Create the bar plot
plt.bar(students, marks)

# Add labels and title
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks Scored by Students")

# Show the plot
plt.show()
```

3. Scatter Plot                                                                  
o Generate two lists of 20 random integers each (for x and y coordinates).       
o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.randint(1, 20, 20)
y = np.random.randint(1, 20, 20)

# Create the scatter plot
plt.scatter(x, y, c='blue', s=50, label='Points')

# Add labels, title, and legend
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.title("Scatter Plot of Random Points")
plt.legend()

# Show the plot
plt.show()
```

4. Histogram                                                                                
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.                                           
o Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)

# Create the histogram
plt.hist(data, bins=10, color='green', edgecolor='black')

# Add labels, title, and grid
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.grid()

# Show the plot
plt.show()
```

5. Pie Chart                                                                     
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']                                                                             
with values [20, 30, 25, 15, 10].                                                   
o Display the percentage values on the chart and use a color palette for
differentiation.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]

# Create the pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Paired.colors)

# Add title
plt.title("Percentage Distribution of Categories")

# Show the plot
plt.show()
```

6. Box Plot (Seaborn)                                                                   
o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].                    
o Highlight the median and any outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

# Create the box plot
sns.boxplot(data=marks)

# Add title
plt.title("Box Plot of Marks")

# Show the plot
plt.show()
```

7. Heatmap (Seaborn)                                                                   
o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.                 
o Add a color bar and annotate the heatmap with the values.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.randint(1, 10, (5, 5))

# Create the heatmap
sns.heatmap(data, annot=True, cmap='coolwarm', cbar=True)

# Add title
plt.title("Heatmap of Random Data")

# Show the plot
plt.show()
```

8. Comparison of Matplotlib and Seaborn                                          
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.                                                     
o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

# Matplotlib bar plot
plt.bar(students, marks, color='blue')
plt.title("Matplotlib Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Seaborn bar plot
sns.barplot(x=students, y=marks, palette='Blues')
plt.title("Seaborn Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

9. Pair Plot (Seaborn)                                                            
o Manually create a small dataset with columns such as 'Age', 'Height',            
'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import seaborn as sns
import pandas as pd

# Create a dataset
data = {
    'Age': [25, 30, 28, 24, 35, 40, 29, 27, 31, 26],
    'Height': [170, 165, 175, 180, 160, 155, 172, 178, 169, 174],
    'Weight': [70, 65, 80, 75, 60, 58, 77, 73, 68, 72]
}

df = pd.DataFrame(data)

# Create the pair plot
sns.pairplot(df)

# Show the plot
plt.show()
```
